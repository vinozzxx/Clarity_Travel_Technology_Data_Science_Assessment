## Step 1 — Data Quality & Missing Values

Before any analysis we must understand and handle missing data to avoid misleading conclusions.

```python
(df.isnull().sum() / len(df)) * 100
```

| Column                | Missing | % Missing | Action           |
| --------------------- | ------- | --------- | ---------------- |
| fare_basis            | 41      | 2.05%     | Fill Mode        |
| ancillary_revenue_inr | 55      | 2.75%     | Fill 0 or Median |
| payment_method        | 104     | 5.2%      | Fill "Unknown"   |
| customer_complaint    | 1752    | 87.6%     | Investigate      |
| satisfaction_score    | 263     | 13.15%    | Median           |


```python

# 1. fare_basis - Missing only 2%.
# Categorical column.- Use most frequent value (Mode).


df["fare_basis"] = df["fare_basis"].fillna(df["fare_basis"].mode()[0])
print(df["fare_basis"].isnull().sum()) 
```

```python
# 2. ancillary_revenue_inr
df["ancillary_revenue_inr"] = df["ancillary_revenue_inr"].fillna(0)
print(df["ancillary_revenue_inr"].describe())
print(df["ancillary_revenue_inr"].isnull().sum()) 
```

```python
# 3. payment_method Categorical.
df["payment_method"] = df["payment_method"].fillna("Unknown")
print(df["payment_method"].isnull().sum()) 
```

```python
# 4. satisfaction_score (Numerical) - Missing = 13%.

df["satisfaction_score"] = df["satisfaction_score"].fillna(df["satisfaction_score"].median())
print(df["payment_method"].isnull().sum()) 

# Why Median? Suppose scores: 1,2,3,4,5 Median = 3
```

```python
# 5. customer_complaint
print(1752 / 2000 * 100)
df["customer_complaint"].value_counts(dropna=False)
```

```python
df["customer_complaint"] = df["customer_complaint"].fillna("No")
```

## Step 8  — Cancellation Patterns

**Q2:** What is the overall cancellation rate? Which factors are most strongly associated?
We visualise lead time, fare amount, booking channel, cabin class, and trip type.

```python
#  Binary cancellation flag 
df['is_cancelled'] = df['booking_status'].isin(['Cancelled', 'Refunded']).astype(int)
cancel_rate = df['is_cancelled'].mean() * 100
print(f"Overall Cancellation / Refund Rate: {cancel_rate:.2f}%")

#  3a. Cancellation by categorical factors 
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, col, title in zip(axes,
    ['trip_type', 'cabin_class', 'booking_channel'],
    ['Trip Type', 'Cabin Class', 'Booking Channel']):
    rates = df.groupby(col)['is_cancelled'].mean().sort_values(ascending=False)
    sns.barplot(x=rates.values * 100, y=rates.index, palette='Set2', ax=ax)
    ax.set_title(f'Cancellation Rate by {title}')
    ax.set_xlabel('Cancellation Rate (%)')
    ax.axvline(cancel_rate, color='red', linestyle='--', linewidth=1.2, label=f'Overall ({cancel_rate:.1f}%)')
    ax.legend(fontsize=8)

plt.suptitle('Cancellation Rates by Key Categorical Factors', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
#  3b. Lead Time vs Cancellation (Visualisation 1) 
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Box plot
sns.boxplot(x='is_cancelled', y='lead_time_days', data=df, palette='Set3', ax=axes[0])
axes[0].set_xticklabels(['Not Cancelled', 'Cancelled/Refunded'])
axes[0].set_title('Lead Time Distribution vs Cancellation Status')
axes[0].set_ylabel('Lead Time (Days)')

# Lead-time bin vs rate
df['lead_time_bin'] = pd.cut(df['lead_time_days'], bins=[0, 7, 30, 90, 180, 365],
                              labels=['0-7d', '8-30d', '31-90d', '91-180d', '181d+'])
bin_rates = df.groupby('lead_time_bin', observed=True)['is_cancelled'].mean() * 100
sns.barplot(x=bin_rates.index, y=bin_rates.values, palette='coolwarm', ax=axes[1])
axes[1].set_title('Cancellation Rate by Lead-Time Bucket')
axes[1].set_ylabel('Cancellation Rate (%)')
axes[1].axhline(cancel_rate, color='red', linestyle='--', linewidth=1.2)

plt.suptitle('Lead Time vs Cancellation Patterns', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 3c. Fare Amount vs Cancellation (Visualisation 2)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Violin plot: fare distribution
sns.violinplot(x='is_cancelled', y='total_fare_inr', data=df, palette='pastel',
               cut=0, ax=axes[0])
axes[0].set_xticklabels(['Not Cancelled', 'Cancelled/Refunded'])
axes[0].set_title('Total Fare Distribution vs Cancellation Status')
axes[0].set_ylabel('Total Fare (₹)')

# Fare quintile vs cancellation rate
df['fare_quintile'] = pd.qcut(df['total_fare_inr'], q=5,
                               labels=['Q1\n(Lowest)', 'Q2', 'Q3', 'Q4', 'Q5\n(Highest)'])
fare_rates = df.groupby('fare_quintile', observed=True)['is_cancelled'].mean() * 100
sns.barplot(x=fare_rates.index, y=fare_rates.values, palette='rocket', ax=axes[1])
axes[1].set_title('Cancellation Rate by Fare Quintile')
axes[1].set_ylabel('Cancellation Rate (%)')
axes[1].axhline(cancel_rate, color='red', linestyle='--', linewidth=1.2)

plt.suptitle('Fare Amount vs Cancellation Patterns', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

#  Quick correlation table 
print("\nPoint-biserial correlation with is_cancelled:")
num_cols = ['lead_time_days', 'total_fare_inr', 'base_fare_inr', 'pax_count', 'prior_bookings']
corr = df[num_cols + ['is_cancelled']].corr()['is_cancelled'].drop('is_cancelled').sort_values(key=abs, ascending=False)
print(corr.round(3).to_string())
```

## Step 9 — Booking Trends

**Q3:** Are there seasonal/monthly booking trends? How does the channel mix change over the year?

```python
df['booking_month'] = df['booking_date'].dt.month
df['booking_month_name'] = df['booking_date'].dt.strftime('%b')
month_order = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

fig, axes = plt.subplots(2, 1, figsize=(14, 10))

#  4a. Monthly volume 
monthly_counts = df.groupby('booking_month_name').size().reindex(month_order)
sns.barplot(x=monthly_counts.index, y=monthly_counts.values, palette='Blues_d', ax=axes[0])
axes[0].set_title('Monthly Booking Volume (2025)', fontsize=13, fontweight='bold')
axes[0].set_ylabel('Number of Bookings')
axes[0].set_xlabel('')

#  4b. Channel mix over months (stacked bar) 
channel_month = df.groupby(['booking_month_name', 'booking_channel']).size().unstack(fill_value=0)
channel_month = channel_month.reindex(month_order)
channel_month_pct = channel_month.div(channel_month.sum(axis=1), axis=0) * 100
channel_month_pct.plot(kind='bar', stacked=True, ax=axes[1],
                       colormap='tab10', edgecolor='white', linewidth=0.5)
axes[1].set_title('Booking Channel Mix by Month (%)', fontsize=13, fontweight='bold')
axes[1].set_ylabel('Share of Bookings (%)')
axes[1].set_xlabel('Month')
axes[1].legend(title='Channel', bbox_to_anchor=(1.01, 1), loc='upper left')
axes[1].set_xticklabels(month_order, rotation=0)

plt.tight_layout()
plt.show()
```

## Step 10 — Customer Behaviour

**Q4:** Compare repeat customers (`prior_bookings > 0`) vs new customers on fare value,
cancellation rate, and satisfaction scores.

```python
df['customer_type'] = np.where(df['prior_bookings'] > 0, 'Repeat Customer', 'New Customer')

summary = df.groupby('customer_type').agg(
    count        = ('booking_id', 'count'),
    avg_fare     = ('total_fare_inr', 'mean'),
    cancel_rate  = ('is_cancelled', 'mean'),
    avg_sat      = ('satisfaction_score', 'mean'),
    avg_prior    = ('prior_bookings', 'mean')
).round(2)
summary['cancel_rate'] = (summary['cancel_rate'] * 100).round(2)
display(summary)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
palette = {'New Customer': '#4c72b0', 'Repeat Customer': '#55a868'}

for ax, col, label in zip(axes,
    ['avg_fare', 'cancel_rate', 'avg_sat'],
    ['Avg Total Fare (₹)', 'Cancellation Rate (%)', 'Avg Satisfaction Score']):
    vals = summary[col]
    sns.barplot(x=vals.index, y=vals.values, palette=palette, ax=ax)
    ax.set_title(label)
    ax.set_xlabel('')

plt.suptitle('New vs Repeat Customers — Key Metrics', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Summary & Top 3 Business Recommendations

### Executive Summary

Analysis of 2,000 airline booking records revealed several important trends in customer behaviour, booking patterns, revenue generation, and operational performance. Long-haul flights accounted for the majority of bookings, while Economy class represented the largest share of passengers. Revenue was concentrated among a small group of airlines and premium cabin classes, with B2B Agent channels contributing a significant portion of overall bookings. The dataset also showed measurable cancellation activity and recurring customer service issues related to refunds, seat allocation, and ancillary services.

### Recommendation 1: Improve Refund and Post-Booking Service Processes

Customer complaints were primarily related to refund delays, ticket servicing, and ancillary product issues. Implementing automated refund tracking, proactive customer notifications, and faster service-resolution workflows could significantly improve customer satisfaction and reduce support workload.

### Recommendation 2: Increase Ancillary Revenue Opportunities

Ancillary services such as baggage, seat selection, and meal purchases represent an additional revenue stream. Clarity should introduce targeted upselling strategies during the booking journey, particularly for long-haul and premium-cabin passengers, who are more likely to purchase add-on services.

### Recommendation 3: Strengthen Direct Digital Channels

A large proportion of bookings originated from B2B Agent channels. Expanding direct customer acquisition through the website and mobile application can reduce dependency on intermediaries, improve customer retention, and provide greater access to customer behaviour data. Loyalty incentives, personalised offers, and mobile-exclusive promotions can help accelerate adoption of direct channels.

Overall, the analysis suggests that operational improvements, ancillary revenue optimisation, and stronger digital engagement initiatives could positively impact both customer experience and business profitability.
